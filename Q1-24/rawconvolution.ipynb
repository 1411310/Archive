```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('valkyrie-left-top.webp')
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

kernel1 = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])
im1 = cv2.filter2D(img, -1, kernel1)
plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(im1, cv2.COLOR_BGR2RGB))
plt.title('Identity Kernel')

kernel2 = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
im2 = cv2.filter2D(img, -1, kernel2)
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.title('Sharpening Kernel')

# Blurring kernel
kernel3 = np.array([[0.02040816, 0.02040816, 0.02040816, 0.02040816, 0.02040816,
                    0.02040816, 0.02040816],
                   [0.02040816, 0.02040816, 0.02040816, 0.02040816, 0.02040816,
                    0.02040816, 0.02040816],
                   [0.02040816, 0.02040816, 0.02040816, 0.02040816, 0.02040816,
                    0.02040816, 0.02040816],
                   [0.02040816, 0.02040816, 0.02040816, 0.02040816, 0.02040816,
                    0.02040816, 0.02040816],
                   [0.02040816, 0.02040816, 0.02040816, 0.02040816, 0.02040816,
                    0.02040816, 0.02040816],
                   [0.02040816, 0.02040816, 0.02040816, 0.02040816, 0.02040816,
                    0.02040816, 0.02040816],
                   [0.02040816, 0.02040816, 0.02040816, 0.02040816, 0.02040816,
                    0.02040816, 0.02040816]])
im3 = cv2.filter2D(img, -1, kernel3)
plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(im3, cv2.COLOR_BGR2RGB))
plt.title('Blurring Kernel')

plt.show()
```
